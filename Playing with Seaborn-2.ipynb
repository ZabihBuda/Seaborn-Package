```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Regression Line
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Let's visualize two regression lines in one chart based on sex by using the 'hue' parameter
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()
```

```python
# we can also have one regression line for each gender using one line of code
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
plt.show()
```

```python
# Residual Plot

sns.residplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Strip Plot
sns.stripplot(x='day', y='tip', data=tips, size=5, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
# Boxplot and Violin Plot
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
 
plt.subplot(1, 2, 2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
# Joint plot
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Joint plot with density estimate
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```python
# Pair plot
sns.pairplot(tips)
plt.show()
```

```python
#Heatmap
df_corr=tips.corr()

sns.heatmap(df_corr)
plt.title('Correlation plot')
plt.show()
```

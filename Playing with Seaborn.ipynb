```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style='darkgrid')
```

```python
f=sns.load_dataset('flights')
f.head()
```

```python
sns.relplot(x='passengers', y='month', data=f)
```

```python
sns.relplot(x='passengers', y='month', data=f, hue='year')
```

```python
sns.relplot(x='passengers', y='month', data=f, hue='year', palette='ch:r=-.5, l=.75')
```

# Line plot Visualizaiton

```python
a=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7])
a.head()
```

```python
sns.relplot(x='Day', y='Clothes', kind='line', data=a)
```

```python
g=sns.relplot(x='Day', y='Clothes', kind='line', data=a)
g.fig.autofmt_xdate()
```

# Categorical Data Visualization

```python
sns.set(style='ticks', color_codes=True)
a=sns.load_dataset('tips')
a.head()
```

```python
sns.catplot(x='day', y='total_bill', data=a)
```

```python
sns.catplot(x='time', y='total_bill', data=a)
```

```python
a.head()
```

```python
sns.catplot(x='sex', y='tip', data=a)
```

```python
sns.catplot(x='sex', y='tip', data=a, kind='violin')
```

# Visualizing the distribution of dataset

```python
import pandas as pd
import numpy as np
from scipy import stats
sns.set(color_codes=True)
```

```python
# Plotting Univariate Distribution
a=np.random.normal(loc=5, size=100, scale=2)
sns.distplot(a)
```

```python
a.head()
```

```python
# Multi-plot Grids
sns.set(style='darkgrid')
a=sns.load_dataset('iris')
b=sns.FacetGrid(a, col='species')
b.map(plt.hist, 'sepal_length')
```

```python
sns.set(style='ticks', color_codes=True)
t=sns.load_dataset('tips')
t.head()

```

```python
t=sns.FacetGrid(t, col='sex')
t.map(plt.hist, 'tip')
```

```python
t=sns.FacetGrid(t, col='time')
t.map(plt.hist, 'tip')
```

```python
# Pair Grid

sns.set(style='ticks')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```
